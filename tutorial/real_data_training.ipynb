```python
localizer = DataLocalizer()

localizer.checkpoint_search_dir = Config.SECOND_CHECKPOINT_SEARCH_DIR
localizer.load_checkpoint = True
localizer.init_ensure_module()
localizer.test_data = test_data
localizer.validate_predictions()
```

```python
from matplotlib.pyplot import eventplot, imshow, figure, xlim, savefig
from tqdm import trange
import os
# print(num_train_sumples)
for i in trange(3):
    segment_3d = test_data[i].molecule.bionano_image.segment_image[0]
    inference_item = localizer.inference_item(segment_3d)
    indices = numpy.flatnonzero(test_data[i].ground_truth.label_target)
    gt = indices + test_data[i].ground_truth.loc_target[indices]
    figure(figsize=(30, 3))
    image_input = segment_3d
    target_width = localizer.image_channels
    source_width = image_input.shape[0] // 2 + 1
    image_input = image_input[source_width - target_width // 2: source_width + target_width // 2 + 1]
    imshow(image_input, aspect="auto", cmap="gray")

    eventplot([gt , inference_item.loc_pred], colors=["b", "r"])
    if not os.path.isdir(Config.EVALUATION_DIR):
            os.makedirs(Config.EVALUATION_DIR)
    savefig(Config.EVALUATION_DIR / f"eval_image_mol_{i}.png")
    
    # eventplot(inference_item.loc_pred)
```
